```python
from typing import Optional

"""
给你一个链表的头节点 head 和一个特定值 x ，请你对链表进行分隔，使得所有 小于 x 的节点都出现在 大于或等于 x 的节点之前。
你应当 保留 两个分区中每个节点的初始相对位置。

示例 1：
输入：head = [1,4,3,2,5,2], x = 3
输出：[1,2,2,4,3,5]

示例 2：
输入：head = [2,1], x = 2
输出：[1,2]

提示：
链表中节点的数目在范围 [0, 200] 内
-100 <= Node.val <= 100
-200 <= x <= 200
"""


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        """
            改进建议：
                不使用 x_node：
                你可以直接使用两个虚拟节点 less_dummy 和 greater_dummy 来分别存储小于 x 和大于或等于 x 的节点，最后将它们连接起来。
            简化逻辑：
                通过两个指针分别处理小于 x 和大于或等于 x 的节点，最后将它们连接起来。
        """
        """
            边界情况：
                    如果链表为空或只有一个节点，直接返回 head 是正确的，但后续的逻辑可能会破坏这种简单情况。
        """
        if not head or not head.next:
            return head
        """
            x_node 的使用：
                    你创建了一个值为 x 的节点 x_node，并将其插入到链表中。然而，这个节点并不是原始链表中的节点，而是你手动插入的。这可能会导致链表结构混乱，尤其是当 x 不在链表中时。
        """
        dummy, x_node = ListNode(), ListNode(x)
        prev = dummy
        prev.next = x_node
        tail = x_node

        curr = head
        while curr:
            #保存下一个节点
            next_node = curr.next
            #判断当前节点和x的值大小
            if curr.val < x:
                #当前节点指向x节点
                curr.next = x_node
                #x节点的先驱节点指向当前节点
                prev.next = curr
                #x的先驱节点向后移动
                prev = prev.next
            else:
                #尾节点指向当前节点
                tail.next = curr
                #尾节点后移
                tail = tail.next
            curr = next_node
        """
            tail.next = None：
                这行代码的目的是确保链表的末尾指向 None，但它的位置可能不正确，可能会导致链表断裂。
        """
        tail.next = None
        """
            prev.next = prev.next.next：
                这行代码的目的是删除 x_node，但它的逻辑不够清晰，可能会导致链表断裂或丢失节点。
        """
        prev.next = prev.next.next
        return dummy.next

    def partition2(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        if not head or not head.next:
            return head

        # 创建两个虚拟节点，分别用于存储小于 x 和大于等于 x 的节点
        less_dummy = ListNode()
        greater_dummy = ListNode()

        # 创建两个指针分别指向两个虚拟节点的末尾
        less_tail = less_dummy
        greater_tail = greater_dummy

        curr = head
        while curr:
            if curr.val < x:
                less_tail.next = curr
                less_tail = less_tail.next
            else:
                greater_tail.next = curr
                greater_tail = greater_tail.next
            curr = curr.next

        # 将小于 x 的链表和大于等于 x 的链表连接起来
        less_tail.next = greater_dummy.next
        # 确保大于等于 x 的链表的末尾指向 None
        greater_tail.next = None

        return less_dummy.next


node1 = ListNode(1)
node2 = ListNode(4)
node3 = ListNode(3)
node4 = ListNode(2)
node5 = ListNode(5)
node6 = ListNode(2)
node7 = ListNode(2)
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6
node6.next = node7

result = Solution().partition(node1, 3)
while result:
    print(result.val)
    result = result.next
        
```
