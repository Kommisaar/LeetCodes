```python
from typing import Optional

"""
给你链表的头节点 head ，每 k 个节点一组进行翻转，请你返回修改后的链表。
k 是一个正整数，它的值小于或等于链表的长度。如果节点总数不是 k 的整数倍，那么请将最后剩余的节点保持原有顺序。
你不能只是单纯的改变节点内部的值，而是需要实际进行节点交换。

示例 1：
输入：head = [1,2,3,4,5], k = 2
输出：[2,1,4,3,5]

示例 2：
输入：head = [1,2,3,4,5], k = 3
输出：[3,2,1,4,5]
"""


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def reverse(head):
    if not head or not head.next:
        return head, head

    prev = None
    tail = head

    while head:
        next_node = head.next
        head.next = prev
        prev = head
        head = next_node

    return prev, tail


class Solution:
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:

        if not head:
            return head
        dummy = prev = ListNode()
        dummy.next = head

        while prev:
            #保存当前组的前驱节点
            curr_pair_prev = prev

            #获取下一组的前驱节点
            cnt = 0
            while prev.next and cnt < k:
                prev = prev.next
                cnt += 1

            #如果不足k个，则结束反转
            if cnt != k:
                return dummy.next

            #获取下一组的头节点
            next_pair_head = prev.next
            #生成反转组的终点
            prev.next = None
            #反转链表组，并返回头节点和尾节点
            reversed_head, reversed_tail = reverse(curr_pair_prev.next)
            #将当前组的先驱节点指向反转列表的头节点
            curr_pair_prev.next = reversed_head
            #将反转列表的尾节点指向下一组列表的头节点
            reversed_tail.next = next_pair_head
            #将当前组的前驱节点移动到下一组的前驱节点
            prev = reversed_tail

        return dummy.next


node1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node5 = ListNode(5)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

solution = Solution()
print()
result = solution.reverseKGroup(node1, 2)
while result:
    print(result.val)
    result = result.next
```
