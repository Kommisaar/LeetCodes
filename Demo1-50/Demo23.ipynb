```python
import heapq
from heapq import heapify
from typing import Optional, List

"""
给你一个链表数组，每个链表都已经按升序排列。
请你将所有链表合并到一个升序链表中，返回合并后的链表。
 
示例 1：
输入：lists = [[1,4,5],[1,3,4],[2,6]]
输出：[1,1,2,3,4,4,5,6]
解释：链表数组如下：
[
  1->4->5,
  1->3->4,
  2->6
]
将它们合并到一个有序链表中得到。
1->1->2->3->4->4->5->6

示例 2：
输入：lists = []
输出：[]

示例 3：
输入：lists = [[]]
输出：[]
"""


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def merge_list(list1, list2):
    dummy = ListNode()  # 创建一个虚拟头节点
    curr = dummy  # 当前指针

    while list1 and list2:
        if list1.val <= list2.val:
            curr.next = list1
            list1 = list1.next
        else:
            curr.next = list2
            list2 = list2.next
        curr = curr.next

    # 将剩余的链表直接连接到结果链表
    curr.next = list1 if list1 else list2

    return dummy.next


class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists:
            return None

        if len(lists) == 1:
            return lists[0]

        if len(lists) == 2:
            return merge_list(lists[0], lists[1])

        if len(lists) > 2:
            mid = len(lists) // 2
            return merge_list(self.mergeKLists(lists[:mid]), self.mergeKLists(lists[mid:]))

    def mergeKLists2(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists:
            return None

        # 最小堆，存储 (node.val, index, node)
        # index 用于解决节点值相同时的比较问题
        heap = []
        for i, head in enumerate(lists):
            if head:
                heapq.heappush(heap, (head.val, i, head))

        # 虚拟头节点
        dummy = ListNode()
        curr = dummy

        while heap:
            val, i, node = heapq.heappop(heap)  # 弹出最小节点
            curr.next = node  # 将最小节点连接到结果链表
            curr = curr.next  # 移动当前指针

            # 如果当前节点有下一个节点，将其推入堆中
            if node.next:
                heapq.heappush(heap, (node.next.val, i, node.next))

        return dummy.next
        
```
