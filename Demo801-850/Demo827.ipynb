```python
import copy
from collections import deque, defaultdict
from typing import List

"""
给你一个大小为 n x n 二进制矩阵 grid 。最多 只能将一格 0 变成 1 。
返回执行此操作后，grid 中最大的岛屿面积是多少？
岛屿 由一组上、下、左、右四个方向相连的 1 形成。

示例 1:
输入: grid = [[1, 0], [0, 1]]
输出: 3
解释: 将一格0变成1，最终连通两个小岛得到面积为 3 的岛屿。

示例 2:
输入: grid = [[1, 1], [1, 0]]
输出: 4
解释: 将一格0变成1，岛屿的面积扩大为 4。

示例 3:
输入: grid = [[1, 1], [1, 1]]
输出: 4
解释: 没有0可以让我们变成1，面积依然为 4。
 
提示：

n == grid.length
n == grid[i].length
1 <= n <= 500
grid[i][j] 为 0 或 1
"""


class Solution:
    def largestIsland(self, grid: List[List[int]]) -> int:
        neighbours = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        max_area, m, n = 0, len(grid), len(grid[0])
        visited = [False] * m * n
        visit_queue = deque([])
        islands_area_table = {}
        island_index = 2
        for i in range(m):
            for j in range(n):

                if grid[i][j] == 1 and not visited[i * n + j]:
                    area = 1
                    visited[i * n + j] = True
                    visit_queue.append((i, j))
                    grid[i][j] = island_index
                    

                    while visit_queue:
                        x, y = visit_queue.popleft()

                        for dx, dy in neighbours:
                            nx, ny = x + dx, y + dy

                            if 0 <= nx < m and 0 <= ny < n and grid[nx][ny] == 1 and not visited[nx * n + ny]:
                                area += 1
                                visited[nx * n + ny] = True
                                visit_queue.append((nx, ny))
                                grid[nx][ny] = island_index
                    
                    islands_area_table[island_index] = area
                    max_area = max(max_area, area)
                    island_index += 1
        
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 0:
                    connected_island = set()
                    for dx, dy in neighbours:
                        nx, ny = i + dx, j + dy
                        if 0 <= nx < m and 0 <= ny < n and grid[nx][ny] in islands_area_table:
                            connected_island.add(grid[nx][ny])
                    area = 1
                    for island in connected_island:
                        print(connected_island)
                        area += islands_area_table[island]
                    max_area = max(max_area, area)
        print(islands_area_table)
        return max_area
    
    def largestIsland2(self, grid: List[List[int]]) -> int:
        neighbours = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        max_area, m, n = 0, len(grid), len(grid[0])
        visited = [[False] * n for _ in range(m)]
        islands_area_table = defaultdict(int)
        island_index = 2
    
        # 第一次遍历，标记所有岛屿并计算面积
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1 and not visited[i][j]:
                    area = 1
                    visited[i][j] = True
                    visit_queue = deque([(i, j)])
                    grid[i][j] = island_index
    
                    while visit_queue:
                        x, y = visit_queue.popleft()
                        for dx, dy in neighbours:
                            nx, ny = x + dx, y + dy
                            if 0 <= nx < m and 0 <= ny < n and grid[nx][ny] == 1 and not visited[nx][ny]:
                                area += 1
                                visited[nx][ny] = True
                                visit_queue.append((nx, ny))
                                grid[nx][ny] = island_index
    
                    islands_area_table[island_index] = area
                    max_area = max(max_area, area)
                    island_index += 1
    
        # 第二次遍历，尝试将每个 0 变成 1，并计算最大面积
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 0:
                    connected_island = set()
                    for dx, dy in neighbours:
                        nx, ny = i + dx, j + dy
                        if 0 <= nx < m and 0 <= ny < n and grid[nx][ny] in islands_area_table:
                            connected_island.add(grid[nx][ny])
                    area = 1
                    for island in connected_island:
                        area += islands_area_table[island]
                    max_area = max(max_area, area)
    
        return max_area
    
    def largestIsland3(self, grid: List[List[int]]) -> int:
        neighbours = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        m, n = len(grid), len(grid[0])
        islands_area_table = {}
        island_index = 2
        max_area = 0
    
        # 第一次遍历，标记所有岛屿并计算面积
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1:
                    area = 1
                    grid[i][j] = island_index
                    visit_queue = deque([(i, j)])
    
                    while visit_queue:
                        x, y = visit_queue.popleft()
                        for dx, dy in neighbours:
                            nx, ny = x + dx, y + dy
                            if 0 <= nx < m and 0 <= ny < n and grid[nx][ny] == 1:
                                area += 1
                                grid[nx][ny] = island_index
                                visit_queue.append((nx, ny))
    
                    islands_area_table[island_index] = area
                    max_area = max(max_area, area)
                    island_index += 1
    
        # 如果没有 0，直接返回整个网格的面积
        if max_area == m * n:
            return max_area
    
        # 第二次遍历，尝试将每个 0 变成 1，并计算最大面积
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 0:
                    connected_island = set()
                    for dx, dy in neighbours:
                        nx, ny = i + dx, j + dy
                        if 0 <= nx < m and 0 <= ny < n and grid[nx][ny] in islands_area_table:
                            connected_island.add(grid[nx][ny])
                    area = 1
                    for island in connected_island:
                        area += islands_area_table[island]
                    max_area = max(max_area, area)
    
        return max_area
    
arr = [[1,1],[1,1]]
arr2 =[[1,0],[0,1]]
arr3 =[[1,1],[1,0]]

print(Solution().largestIsland3(arr3))
```
