```python
from collections import deque
from typing import Optional, List

"""
给定一个 N 叉树，找到其最大深度。
最大深度是指从根节点到最远叶子节点的最长路径上的节点总数。
N 叉树输入按层序遍历序列化表示，每组子节点由空值分隔（请参见示例）。

示例 1：
输入：root = [1,null,3,2,4,null,5,6]
输出：3

示例 2：
输入：root = [1,null,2,3,4,5,null,null,6,7,null,8,null,9,10,null,null,11,null,12,null,13,null,null,14]
输出：5

提示：

树的深度不会超过 1000 。
树的节点数目位于 [0, 104] 之间。
"""


# Definition for a Node.
class Node:
    def __init__(self, val: Optional[int] = None, children: Optional[List['Node']] = None):
        self.val = val
        self.children = children


class Solution:
    def maxDepth(self, root: 'Node') -> int:
        if not root:
            return 0
        return 1 + max(self.maxDepth(i) for i in root.children) if root.children else 1
    
    def maxDepth2(self, root: 'Node') -> int:
        """BFS+队列遍历"""
        if not root:
            return 0
        
        queue = deque([root])  # 初始化队列，根节点入队
        depth = 0  # 记录深度
        
        while queue:
            depth += 1  # 每遍历一层，深度加 1
            level_size = len(queue)  # 当前层的节点数
            for _ in range(level_size):
                node = queue.popleft()  # 弹出当前层的节点
                for child in node.children:  # 将当前节点的所有子节点入队
                    queue.append(child)
        
        return depth
    
    def maxDepth3(self, root: 'Node') -> int:
        """DFS+栈遍历"""
        if not root:
            return 0
        
        stack = [(root, 1)]  # 初始化栈，存储节点和当前深度
        max_depth = 0  # 记录最大深度
        
        while stack:
            node, depth = stack.pop()  # 弹出栈顶节点和当前深度
            max_depth = max(max_depth, depth)  # 更新最大深度
            for child in node.children:  # 将子节点和深度加 1 入栈
                stack.append((child, depth + 1))
        
        return max_depth
```
