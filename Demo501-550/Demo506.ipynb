```python
from typing import List

"""
给你一个长度为 n 的整数数组 score ，其中 score[i] 是第 i 位运动员在比赛中的得分。所有得分都 互不相同 。

运动员将根据得分 决定名次 ，其中名次第 1 的运动员得分最高，名次第 2 的运动员得分第 2 高，依此类推。运动员的名次决定了他们的获奖情况：

名次第 1 的运动员获金牌 "Gold Medal" 。
名次第 2 的运动员获银牌 "Silver Medal" 。
名次第 3 的运动员获铜牌 "Bronze Medal" 。
从名次第 4 到第 n 的运动员，只能获得他们的名次编号（即，名次第 x 的运动员获得编号 "x"）。
使用长度为 n 的数组 answer 返回获奖，其中 answer[i] 是第 i 位运动员的获奖情况。

示例 1：
输入：score = [5,4,3,2,1]
输出：["Gold Medal","Silver Medal","Bronze Medal","4","5"]
解释：名次为 [1st, 2nd, 3rd, 4th, 5th] 。

示例 2：
输入：score = [10,3,8,9,4]
输出：["Gold Medal","5","Bronze Medal","Silver Medal","4"]
解释：名次为 [1st, 5th, 3rd, 2nd, 4th] 。
"""


class Solution:
    class MaxHeap:
        def __init__(self):
            self.heap = []

        def parent(self, i):
            return (i - 1) // 2

        def lchild(self, i):
            return i * 2 + 1

        def rchild(self, i):
            return i * 2 + 2

        def swap(self, i, j):
            self.heap[i], self.heap[j] = self.heap[j], self.heap[i]

        def heap_up(self, i):
            while i > 0 and self.heap[self.parent(i)][1] < self.heap[i][1]:
                parent = self.parent(i)
                self.swap(i, parent)
                i = parent

        def heap_down(self, i):
            n = len(self.heap)
            largest, lchild, rchild = i, self.lchild(i), self.rchild(i)
            if lchild < n and self.heap[lchild][1] > self.heap[largest][1]:
                largest = lchild
            if rchild < n and self.heap[rchild][1] > self.heap[largest][1]:
                largest = rchild

            if i != largest:
                self.swap(i, largest)
                self.heap_down(largest)

        def extract_max(self):
            if not self.heap:
                return None
            max_val = self.heap[0]
            self.heap[0] = self.heap[-1]
            self.heap.pop()
            self.heap_down(0)
            return max_val

        def build_max_heap(self, arr):
            self.heap = [(i, arr[i]) for i in range(len(arr))]
            n = len(arr)
            for i in range(n // 2 - 1, -1, -1):
                self.heap_down(i)

    def findRelativeRanks(self, score: List[int]) -> List[str]:
        prize_dict = {1: 'Gold Medal', 2: 'Silver Medal', 3: 'Bronze Medal'}
        score_heap = Solution.MaxHeap()
        score_heap.build_max_heap(score)
        res = [-1] * len(score)
        rank = 1
        while score_heap.heap:
            index = score_heap.extract_max()[0]
            res[index] = prize_dict.get(rank, str(rank))
            rank += 1

        return res

    def findRelativeRanks2(self, score: List[int]) -> List[str]:
        rank_to_prize = {1: 'Gold Medal', 2: 'Silver Medal', 3: 'Bronze Medal'}
        score_dict = {}
        for i in range(len(score)):
            score_dict[score[i]] = i
        score.sort(reverse=True)
        res = [-1]*len(score)
        for i in range(len(score)):
            res[score_dict[score[i]]] = rank_to_prize.get(i+1,str(i+1))
        
        return res
        


print(Solution.findRelativeRanks2(Solution(), [2, 1, 4, 5, 6, 8, 9]))
            
```
